```python
from tkinter import *
import datetime
import time
import winsound
import threading

# Global variable to track if an alarm is active
alarm_active = False

def alarm(set_alarm_timer):
    global alarm_active
    while alarm_active:
        time.sleep(1)
        current_time = datetime.datetime.now()
        now = current_time.strftime("%H:%M:%S")
        print("Current Time:", now)
        if now >= set_alarm_timer:
            print("Time to Wake up!")
            for _ in range(5):
                winsound.Beep(1000, 500)
                time.sleep(0.5)
            alarm_active = False
            break

def actual_time():
    global alarm_active
    alarm_active = False
    time.sleep(1.1)  # Ensure previous thread stops
    
    try:
        h = int(hour.get())
        m = int(min.get())
        s = int(sec.get())

        if h < 0 or h > 23 or m < 0 or m > 59 or s < 0 or s > 59:
            raise ValueError("Invalid time range")
        
        alarm_time = f"{h:02}:{m:02}:{s:02}"
        alarm_active = True
        threading.Thread(target=alarm, args=(alarm_time,), daemon=True).start()
        print(f"Alarm set for {alarm_time}")
        
    except ValueError as e:
        print(f"Error: {e}. Please enter valid numbers (HH:MM:SS).")

clock = Tk()
clock.title("Alarm Clock")
clock.geometry("400x200")

Label(clock, text="Enter time in 24 hour format!", fg="red", bg="black", font="Arial").place(x=60, y=150)
Label(clock, text="Hour  Min   Sec", font=60).place(x=110, y=20)
Label(clock, text="When to wake you up", fg="blue", relief="solid", font=("Helvetica", 10, "bold")).place(x=100, y=0)

# Variables for time input
hour = StringVar()
min = StringVar()
sec = StringVar()

Entry(clock, textvariable=hour, bg="pink", width=5).place(x=110, y=50)
Entry(clock, textvariable=min, bg="pink", width=5).place(x=170, y=50)
Entry(clock, textvariable=sec, bg="pink", width=5).place(x=230, y=50)

Button(clock, text="Set Alarm", fg="red", width=10, command=actual_time).place(x=150, y=90)

clock.mainloop()
```
